# Train Set

```python
import pandas as pd
traindf = pd.read_csv (r'/home/yupaporn/codes/USAI/traindf.csv')
traindf
```

```python
import os
os.chdir('/media/tohn/SSD/Image/')
!ls
```

```python
import shutil
dic = {"filename": []}
case = list()

path_train = traindf['Path Crop'].tolist() 
#Train
train_Nor_dir = '/media/tohn/SSD/Image/train'
fnames = path_train  
for fname in fnames:
    dst = os.path.join(train_Nor_dir, os.path.basename(fname))
    if os.path.exists(dst):
        dst = dst+'cp.jpg'
    a = dst.split('/')[-1]
    dic["filename"].append(a)
    shutil.copyfile(fname, dst)
```

```python
dics = pd.DataFrame(dic)
len(dics)
```

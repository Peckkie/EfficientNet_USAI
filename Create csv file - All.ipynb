```python
#complete 
Table = list()
# Abs_df = list()

arange =['1 ABNORMAL', '2 ABNORMAL', '3 ABNORMAL', '4 ABNORMAL', '5 ABNORMAL', '6 ABNORMAL', '7 ABNORMAL', '9 ABNORMAL', '8.1 ABNORMAL','8.2,8.3 ABNORMAL', '10 ABNORMAL', '11 ABNORMAL']
for i in arange:
    
    
    if i =='8.2,8.3 ABNORMAL':
        aranges = ['8-2 ABNORMAL', '8-3 ABNORMAL']
        for m in aranges:
            Opath = '/media/tohn/HDD/VISION_dataset/USAI/ABnormal01/8.2,8.3 ABNORMAL/'+m+'/'
            # Spath = Opath+'cropped'
            os.chdir('/media/tohn/HDD/VISION_dataset/USAI/ABnormal01/8.2,8.3 ABNORMAL/'+m+'/') 
            
            imglist = glob.glob("*.JPG")
            for j in imglist :
                A = j.replace('  ',' ')
                a = A.split(' ')
                #subclass
                a_subclass =a[0]
                #Abs
                a_absposi=a[1][0:2]
                #Sub
                if len(a[1])==2:
                    a_sunposi = a[1]
                    if len(a_sunposi) == 7:
                        a_sunposi = 'P41'
                else :
                    a1= a[1].split('-')
                    a_sunposi = a1[0]+a1[1]
                    if len(a_sunposi) == 7:
                        a_sunposi = 'P41'
                #Case
                a_case=a[2][1:4]
                #Class
                a_class = 'Abnormal'
                #Path full images
                a_pathfull = Opath+j
                #path cropped images
                if os.path.exists(Opath+'cropped/'+j) ==True:
                    a_pathcrop = Opath+'cropped/'+j
                else :
                    a_pathcrop = 'None'
    
                Table.append({'Case':a_case,'Abs Position':a_absposi,'Sub Position': a_sunposi, 'Class':a_class, 'Sub_class':a_subclass,'Path Full':a_pathfull,'Path Crop':a_pathcrop})
    
    else :
        Opath = '/media/tohn/HDD/VISION_dataset/USAI/ABnormal01/'+i+'/'
        # Spath = Opath+'cropped'
        os.chdir('/media/tohn/HDD/VISION_dataset/USAI/ABnormal01/'+i+'/')
    
    
        imglist = glob.glob("*.JPG")
        for j in imglist :
            A = j.replace('  ',' ')
            a = A.split(' ')
            #subclass
            a_subclass =a[0]
            #Abs
            a_absposi=a[1][0:2]
            #Sub
            if len(a[1])==2:
                a_sunposi = a[1]
                if len(a_sunposi) == 7:
                    a_sunposi = 'P41'
            else :
                a1= a[1].split('-')
                a_sunposi = a1[0]+a1[1]
                if len(a_sunposi) == 7:
                    a_sunposi = 'P41'
            #Case
            if len(a[2]) == 4:
                if i == '9 ABNORMAL':
                    a_case = a[2][1:4]
                    if a_case =='JPG':
                        a_case = a[1][5:8]
                else :
                    a_case = a[1][5:8]
            else:
                a_case=a[2][1:4]
            #Class
            a_class = 'Abnormal'
            #Path full images
            a_pathfull = Opath+j
            #path cropped images
            if os.path.exists(Opath+'cropped/'+j) ==True:
                a_pathcrop = Opath+'cropped/'+j
            else :
                a_pathcrop = 'None'
            
            
            Table.append({'Case':a_case,'Abs Position':a_absposi,'Sub Position': a_sunposi, 'Class':a_class, 'Sub_class':a_subclass,'Path Full':a_pathfull,'Path Crop':a_pathcrop})
        
```
